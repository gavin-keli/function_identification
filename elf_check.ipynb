```python
binary_path = "./data/lookback-corpus/testsuite-utils-gcc/O0/unstrip/libcrypto.so-amd64-gcc-O0"

with open(binary_path, "rb") as binary_file:
    binary_elf = ELFFile(binary_file)
    #print(binary_elf)

    # Extract the code from the binary.
    data = binary_elf_class._generate_data(binary_elf)
    print(data, len(data))

    # Extract the tags of each byte in the binary code (1 if it is a start of a function, 0 otherwise).
    tags = binary_elf_class._generate_tags(binary_elf)
    print(tags, len(tags))

    symbol_table = binary_elf.get_section_by_name(".symtab")
    #print(symbol_table)

    counter = 0
    start_list = []
    

    for symbol in symbol_table.iter_symbols():
        if symbol["st_info"]["type"] == "STT_FUNC" and symbol["st_size"] != 0:
            counter += 1
            #print(hex(symbol["st_value"]))
            #if hex(symbol["st_value"]) not in start_list:
            start_list.append(symbol["st_value"])
            #print(symbol.entry)
    print(counter)
    print(len(start_list))
    start_list.sort()
    for i in start_list:
        print(i)
```
